```python
import pandas as pd
```

```python
df=pd.read_csv("corona_dat.csv")
```

```python
df.head()
```

```python
df2=df[['date','China','Italy','Spain','Brazil','United Kingdom','US','France','Turkey']]
```

```python
df2.set_index('date',inplace=True) # tarihi index olarak yapmak
```

```python
total=df2.cumsum(axis=0) #bigÃ¼n Ã¶ncesini de ekleyerek gitmesini saÄŸlÄ±yor
```

```python
import bar_chart_race as bcr
```

```python
bcr.bar_chart_race(total,filename="covid.mp4",figsize=(8,5),title="Nurbanu")
```
